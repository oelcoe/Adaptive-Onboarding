````python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import numpy as np

NOTEBOOK_DIR = Path.cwd()
if (NOTEBOOK_DIR / "_sweep_helpers.py").exists():
    sys.path.append(str(NOTEBOOK_DIR))
elif (NOTEBOOK_DIR / "notebooks" / "_sweep_helpers.py").exists():
    sys.path.append(str(NOTEBOOK_DIR / "notebooks"))
else:
    sys.path.append(str(Path(r"C:/Users/49160/Adaptive-Onboarding/notebooks")))

from _sweep_helpers import (
    POLICY_STYLE,
    config_label,
    equalize_y_axes,
    load_run,
    load_sweep,
    plot_metric_lines,
    plot_pareto,
    plot_weighted_delta,
    save,
    setup,
    show_sweeps,
    style_ax,
)

setup()
````

## Select Sweep

````python
PARAM = "dropout"
SWEEP_FILE = None  # paste an exact high_trait_tail sweep filename here
CONFIG_FILTER = {"sensitivity_assignment": "high_trait_tail"}

show_sweeps(PARAM)
sweep = load_sweep(PARAM, sweep_file=SWEEP_FILE, config_filter=CONFIG_FILTER)
print(config_label(sweep))
````

## Safety and Trait-Specific Error

````python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
plot_metric_lines(
    axes[0], sweep, "dropout_rate",
    x_scale=100, y_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Episode dropout rate (%)",
    title="Dropout rate",
)
plot_metric_lines(
    axes[1], sweep, "sensitive_rate",
    x_scale=100, y_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Sensitive questions / questions asked (%)",
    title="Sensitive exposure rate",
)
plot_metric_lines(
    axes[2], sweep, "mean_sensitive_trait_error",
    x_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Mean absolute error on sensitive axes",
    title="Sensitive-axis trait error",
)
fig.suptitle(config_label(sweep), y=1.03, fontsize=9)
plt.tight_layout()
# save(fig, "fig_05_05_trait_tail_safety_error")
plt.show()
````

## Information Quality

````python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
plot_metric_lines(
    axes[0], sweep, "mean_final_d_error",
    x_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Mean D-error",
    title="D-error, all episodes",
)
plot_metric_lines(
    axes[1], sweep, "mean_final_d_error_completed",
    x_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Mean D-error",
    title="D-error, completed only",
)
equalize_y_axes(axes[0], axes[1])
plot_metric_lines(
    axes[2], sweep, "mean_estimation_error",
    x_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Mean estimation error (lower is better)",
    title="Overall estimation error",
)
fig.suptitle(config_label(sweep), y=1.03, fontsize=9)
plt.tight_layout()
# save(fig, "fig_05_05_trait_tail_information")
plt.show()
````

## Trait-Specific Risk vs Error and Weighted Delta

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
plot_pareto(
    axes[0], sweep,
    y_metric="mean_sensitive_trait_error",
    ylabel="Sensitive-axis trait error (lower is better)",
    title="Dropout-risk vs sensitive-axis error",
)
plot_weighted_delta(
    axes[1], sweep, "dropout_rate",
    x_scale=100, y_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Percentage-point delta",
    title="Delta dropout rate",
)
plot_weighted_delta(
    axes[2], sweep, "mean_sensitive_trait_error",
    x_scale=100,
    xlabel="Per-question dropout probability (%)",
    ylabel="Weighted - unweighted",
    title="Delta sensitive-axis error",
)
fig.suptitle(config_label(sweep), y=1.03, fontsize=9)
plt.tight_layout()
# save(fig, "fig_05_05_trait_tail_tradeoff")
plt.show()
````
